```python
import pandas as pd
import numpy as np
```

```python
#Read in txt file of training and testing record
train_df_occ = pd.read_csv("datatraining.txt")
test_df_occ = pd.read_csv("datatest.txt")

#Dropping date attribute 
train_df_occ = train_df_occ.drop(columns='date')
test_df_occ = test_df_occ.drop(columns='date')

#View briefly train dataset
train_df_occ.head()
```

```python
#Shuffling of data
train_df_occ = train_df_occ.sample(frac=1)
test_df_occ = test_df_occ.sample(frac=1)
```

```python
# Export the vector training and test data with label class to csv, the shuffled version every time run will be different, as backup
train_df_occ.to_csv('train_df_occ.csv', index=False)
test_df_occ.to_csv('test_df_occ.csv', index=False)
```
