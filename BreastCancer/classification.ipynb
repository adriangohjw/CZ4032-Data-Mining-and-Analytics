# **DATA PREPROCESSING**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('data.csv')
data.head()
```

```python
data.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
data['diagnosis_enc']=LabelEncoder().fit_transform(data[['diagnosis']])
data.head()
```

```python
data.drop('diagnosis',axis=1,inplace=True)
data.drop('id',axis=1,inplace=True)
data.drop(['Unnamed: 32'],axis=1,inplace=True)
data.head()
```

```python
#getting the correlation values  to our target variable - diagnosis_enc
cor_target=abs(data.corr()['diagnosis_enc'])

#view all the feature columns values with a correlation to diagnosis column
cor_target
```

```python
#filtering features that have a strong correlation to our target column i.e >5
cor_target[cor_target>0.5]
```

```python
#get to view our our diagnosis column values are distributed
data.diagnosis_enc.value_counts()
```

## Prepare train and test dataset

```python
#creating our X feature variables and y outcome variable
#X = data.drop('diagnosis',axis=1)
X = data.drop('diagnosis_enc',axis=1)
y = data['diagnosis_enc']
```

```python
#splitting our data into training and testing data set
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

# **Random Forest Classification**

### train_array(X_train) = train dataset WITHOUT occupancy

### test_array(X_test) = test dataset WITHOUT occupancy
### occlabel_array(y_train) = 'Occupancy' label from train dataset

```python
#Random Forest Classification
from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier()
clf.fit(X_train,y_train)
```

```python
#To predict occupancy using the test's set features
clf_pred=clf.predict(X_test)
print(clf_pred)
```

```python
# Transfer the array of predicted popularity of test data into dataframe
clf_pred_df = pd.DataFrame(data=clf_pred, columns=['Diagnosis_enc'])
clf_pred_df.head(10)
```

```python
RFprob_each = clf.predict_proba(X_test)
RFprob_each
```

```python
pos_prob = []
for inner in RFprob_each:
    pos_prob.append(inner[1])
pos_prob[0:10]
```

```python
X_testdf = X_test.copy()
X_testdf.insert(len(X_testdf.columns),'y_test diagnosis_enc',y_test)
X_testdf=X_testdf.reset_index(drop=True)
X_testdf
```

```python
# Insert result, check the prediction correctness
X_testdf.insert(len(X_testdf.columns), 'RForest', clf_pred_df['Diagnosis_enc'])
X_testdf
```

# **Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB

gaus = GaussianNB()
gaus.fit(X_train, y_train)
```

```python
# To predict popularity using test set's features
NB_pred= gaus.predict(X_test)
print(NB_pred)
```

```python
# Transfer the array of predicted popularity of test data into dataframe
NB_pred_df = pd.DataFrame(data=NB_pred, columns=['Diagnosis_enc'])
NB_pred_df.head(10)
```

```python
NBprob_each = gaus.predict_proba(X_test)
NBprob_each
```

```python
pos_prob = []
for inner in NBprob_each:
    pos_prob.append(inner[1])
pos_prob[0:10]
```

```python
# Insert result, check the prediction correctness
X_testdf.insert(len(X_testdf.columns), 'NBGaus', NB_pred_df['Diagnosis_enc'])
X_testdf
```

# **KNN Classification**

```python
from sklearn.neighbors import KNeighborsClassifier 

knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)
```

```python
# To predict popularity using test set's features
KNN_pred= knn.predict(X_test)
print(KNN_pred)
```

```python
# Transfer the array of predicted popularity of test data into dataframe
KNpredct_df = pd.DataFrame(data=KNN_pred, columns=['Diagnosis_enc'])
KNpredct_df.head(10)
```

```python
KNprob_each = knn.predict_proba(X_test)
KNprob_each
```

```python
pos_prob = []
for inner in KNprob_each:
    pos_prob.append(inner[1])
pos_prob[0:10]
```

```python
# Insert result, check the prediction correctness
X_testdf.insert(len(X_testdf.columns), 'KNeigh', KNpredct_df['Diagnosis_enc'])
X_testdf
```

# **Decision Tree**

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
#To predict occupancy using the test's set features
model_pred=model.predict(X_test)
print(model_pred)
```

```python
# Transfer the array of predicted popularity of test data into dataframe
model_pred_df = pd.DataFrame(data=model_pred, columns=['Diagnosis_enc'])
model_pred_df.head(10)
```

```python
DTprob_each = model.predict_proba(X_test)
DTprob_each
```

```python
pos_prob = []
for inner in DTprob_each:
    pos_prob.append(inner[1])
pos_prob[0:10]
```

```python
# Insert result, check the prediction correctness
X_testdf.insert(len(X_testdf.columns), 'DecisionT', clf_pred_df['Diagnosis_enc'])
X_testdf
```

# **Evaluation of Classifiers**

```python
#Import relevant metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
```

```python
#RANDOM FOREST
acRF=accuracy_score(y_test,clf_pred)
fscoreRF=f1_score(y_test,clf_pred)
conmatrixRF = confusion_matrix(y_test,clf_pred)

#NAIVE BAYES
acNB=accuracy_score(y_test,NB_pred)
fscoreNB = f1_score(y_test,NB_pred)
conmatrixNB = confusion_matrix(y_test,NB_pred)

#K NEIGHBOURS
acKNN=accuracy_score(y_test,KNN_pred)
fscoreKNN=f1_score(y_test,KNN_pred)
conmatrixKNN = confusion_matrix(y_test,KNN_pred)

#DECISION TREE
acDTree = accuracy_score(y_test,model_pred)
fscoreDTree = f1_score(y_test,model_pred)
conmatrixDTree = confusion_matrix(y_test,model_pred)
```

```python
df = pd.DataFrame(
    {
        "KNeigh": [acKNN, fscoreKNN, conmatrixKNN],
        "Naive Bayes": [acNB, fscoreNB, conmatrixNB],
        "Random Forest": [acRF, fscoreRF, conmatrixRF],
        "Decision Tree": [acDTree, fscoreDTree, conmatrixDTree]
    },
    columns=["KNeigh", "Naive Bayes", "Random Forest", "Decision Tree"],
    index = ["Accuracy", "Fscore", "Confusion Matrix"])
df

```
